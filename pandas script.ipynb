```python
def splitting(df):
    con=[]
    cat=[]
    for col in df.columns:
        if df[col].dtype=="object":
            cat.append(col)
        elif df[col].dtype=="int64"or"float64"or"int32"or"float32":
            con.append(col)
    df_cat=df.drop(columns=con)
    df_con=df.drop(columns=cat)
    return df_cat,df_con
df_cat,df_con=splitting(df)
```
